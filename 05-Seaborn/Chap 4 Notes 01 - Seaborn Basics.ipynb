```python
#rugplot
sns.rugplot(x='salary', data = df, height=0.5) # almost a 1D scatter plot
```

```python
sns.set(style='darkgrid')
sns.displot(x='salary', data = df, bins=25, kde='True', rug='True') 
# recommended
```

```python
sns.kdeplot(data=df, x='salary')
```

```python
df = pd.read_csv('StudentsPerformance.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(10,7), dpi=200)
sns.boxplot(data=df, y='race/ethnicity', x = 'math score',
           hue = 'test preparation course')
plt.legend(bbox_to_anchor=(1.05, 0.5))
```

```python
plt.figure(figsize=(10,7), dpi=200)
sns.violinplot(data=df, y='race/ethnicity', x = 'math score',
              hue='test preparation course', split=True, inner='quartile',
              palette='Set2', bw=0.1)
plt.legend(loc=(1.05, 0.5))
```

```python
plt.figure(figsize=(10,7), dpi=200)
sns.swarmplot(data=df, x = 'math score', y='race/ethnicity', 
              hue='test preparation course', dodge=True, size=2)
```

```python
plt.figure(figsize=(10,7), dpi=200)
sns.boxenplot(data=df, x = 'math score', y='race/ethnicity', 
              hue='test preparation course')
```

Comparison plot

```python
df.head()
```

```python
plt.figure(figsize=(10,7), dpi=200)
sns.jointplot(data=df, x='math score', y='reading score',
              hue='gender')
```

```python
plt.figure(dpi=200)
sns.pairplot(data=df, hue='gender', corner=True)
# be careful, super large dataframe
# can use pairplot as a first check, if interesting correlation occurs,
# then delve deeper by creating a joint plot
```

```python
# grid plot
```

```python
sns.catplot(data=df, x='gender', y='math score', kind='violin',
           col='lunch', row='test preparation course')
```

```python
plt.figure(dpi=200)
g = sns.PairGrid(data=df, hue='gender')
g = g.map_upper(sns.scatterplot)
g = g.map_lower(sns.kdeplot)
g = g.map_diag(sns.kdeplot)
g = g.add_legend()
```
