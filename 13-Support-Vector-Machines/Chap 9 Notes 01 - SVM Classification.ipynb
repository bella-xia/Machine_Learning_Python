```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../DATA/mouse_viral_study.csv')
```

```python
df.head()
```

```python
sns.scatterplot(x='Med_1_mL', y='Med_2_mL', hue='Virus Present', data=df)

# HYPERPLANE (2d Line)
x = np.linspace(0, 10, 100)
m = -1
b = 11
y = m*x + b
plt.plot(x, y, color='black')
```

```python
from sklearn.svm import SVC
```

```python
help(SVC)
```

```python
X = df.drop('Virus Present', axis=1)
y = df['Virus Present']
```

```python
model = SVC(kernel='linear', C=1000)
```

```python
model.fit(X, y)
```

```python
from svm_margin_plot import plot_svm_boundary
```

```python
plot_svm_boundary(model, X, y)
```

```python
model = SVC(kernel='linear', C=0.005)
model.fit(X, y)
```

```python
plot_svm_boundary(model, X, y)
```

```python
model = SVC(kernel='rbf', C=1, gamma=2)
# gamma determines how much impact a single example training has
# auto = 1 / n_element
model.fit(X, y)
```

```python
plot_svm_boundary(model, X, y)
```

```python
model = SVC(kernel='sigmoid')
model.fit(X, y)
plot_svm_boundary(model, X, y)
# a bad model bc even points in the middle are used 
# to determine the margin
```

```python
model = SVC(kernel='poly', C=0.05, degree=5)
# don't see that much changes despite the increase in degree
# bc a first-degree function is already good enough for the model,
# making all the other degrees insignificant
model.fit(X, y)
plot_svm_boundary(model, X, y)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
svm = SVC()
param_grid = {'C':[0.01, 0.1, 1], 'kernel':['linear', 'rbf']}
```

```python
grid = GridSearchCV(svm, param_grid)
```

```python
grid.fit(X, y)
```
