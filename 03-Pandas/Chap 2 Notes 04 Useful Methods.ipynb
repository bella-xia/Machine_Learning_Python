```python
def yelp(price):
    if price < 10:
        return '$'
    elif price < 20:
        return '$$'
    else:
        return '$$$'
```

```python
df['total_bill'].apply(yelp)
```

```python
def quality(total_bill, tip):
    if tip / total_bill > 0.15:
        return 'Generous'
    else:
        return 'Other'
```

```python
quality(16.99, 1.01)
```

```python
df[['total_bill', 'tip']].apply(lambda df: quality(df['total_bill'], df['tip']), axis = 1)
```

```python
df['Quality'] = np.vectorize(quality)(df['total_bill'], df['tip'])
```
