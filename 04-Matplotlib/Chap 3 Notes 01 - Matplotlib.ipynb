```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(0, 10)
```

```python
y = 2 * x
```

```python
x
```

```python
y
```

```python
plt.plot(x, y)
plt.title('String Title')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.xlim(0, 6)
plt.ylim(0, 12)
plt.savefig('myfirstplot.png')
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, y)
```

```python
a = np.linspace(0, 10, 11)
```

```python
a
```

```python
b = a ** 4
```

```python
b
```

```python
x
```

```python
y
```

```python
fig = plt.figure(dpi= 100, figsize=(2, 2))

# LARGE AXES
axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(a, b)
axes1.set_xlim(0, 8)
axes1.set_ylim(0, 8000)
axes1.set_xlabel('x axis')
axes1.set_ylabel('y axis')
axes1.set_title('Power of Four')
#SMALL AXES
axes2 = fig.add_axes([0.2, 0.5, 0.25, 0.25])
axes2.plot(a, b)
axes2.set_xlim(1, 2)
axes2.set_ylim(0, 50)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoomed in')
fig.savefig('new_figure.png', bbox_inches='tight')
plt.show()
```

```python
type(fig)
```

```python
fig, axes = plt.subplots(nrows=3, ncols=1)
fig.suptitle('Figure Title', fontsize = 16)
for axe in axes:
    axe.plot(x,y)
    axe.set_xlabel('x')
    axe.set_ylabel('y')
fig.subplots_adjust(wspace=1, hspace = 1)
#plt.tight_layout()
fig.savefig('new_subplots.png', bbox_inches='tight')
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

ax.plot(a,a, label='X vs X', color='lightblue', lw=1.2, ls = '-.', marker = 'o')
lines = ax.plot(a, a**2, label='X vs X^2', color='purple', linewidth=2) 
lines[0].set_dashes([1, 1, 2, 3, 4]) 
ax.plot(a,a+1, label='X vs X+1',marker = '+')
# one point dash, one white space, 2 point dash, 3 white spaces, 4 point dash
#can also use RBG HEX code for color
#linestyle: '--', '-', '-.', ':', 'steps'
ax.legend(loc=(-0.3, 0.5))
```
