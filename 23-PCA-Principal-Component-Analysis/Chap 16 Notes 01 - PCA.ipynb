```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../DATA/cancer_tumor_data_features.csv')
```

```python
df.head()
```

```python
sns.heatmap(df)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X = scaler.fit_transform(df)
```

```python
scaled_X.mean(axis=0)
```

```python
covariance_matrix = np.cov(scaled_X, rowvar=False)
```

```python
eigen_values, eigen_vectors = np.linalg.eig(covariance_matrix)
```

```python
eigen_values
```

```python
eigen_vectors
```

```python
# PCA N Feature Space ---> N PC Space
# HOw many PC to choose? N=2
```

```python
num_component =2 
```

```python
sorted_key = np.argsort(eigen_values)[::-1][:num_component]
# sort values --> reverse it --> grab component
```

```python
eigen_values, eigen_vectors = eigen_values[sorted_key],eigen_vectors[:,sorted_key]
```

```python
# Original data --> project --> eigen vectors
principal_components = np.dot(scaled_X, eigen_vectors)
```

```python
from sklearn.decomposition import PCA
```

```python
# help(PCA)
```

```python
pca_model = PCA(n_components=2)
```

```python
pc_result= pca_model.fit_transform(scaled_X)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(pc_result[:, 0], pc_result[:, 1],
           c = cancer_dictionary['target'])
```

```python
pca_model.components_  # directions of maximum variance
```

```python
df_comp = pd.DataFrame(pca_model.components_, 
                       index=['PC1', 'PC2'],
                      columns=df.columns)
```

```python
df_comp
```

```python
plt.figure(figsize=(20, 3), dpi=150)
sns.heatmap(df_comp, annot=True)
```

```python
pca_model.explained_variance_ratio_
```

```python
# the first component explains 44% of the data,
# whereas the second coponeng explains 19% of the data
```

```python
np.sum(pca_model.explained_variance_ratio_)
```

```python
# a total of 63% of the data is explained after dimentional
# reduction
```

```python
pca_30 = PCA(n_components=30)
```

```python
pca_30.fit(scaled_X)
```

```python
pca_30.explained_variance_ratio_
```

```python
np.sum(pca_30.explained_variance_ratio_)
# no reduction occurs, bc from 30 features to 30 components
# do not really do dimentional reduction
```

```python
explained_variance = []

for i in range(1, 30):
    pca = PCA(n_components=i)
    pca.fit(scaled_X)
    explained_variance.append(np.sum(pca.explained_variance_ratio_))
```

```python
plt.plot(range(1, 30), explained_variance)
plt.xlabel('Num of Components')
plt.ylabel('Variance Explained')
```
